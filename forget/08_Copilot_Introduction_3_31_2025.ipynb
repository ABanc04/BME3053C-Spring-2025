# What is GitHub Copilot?

GitHub Copilot is an AI pair programmer that helps you write code faster and with less effort. It works by:

- Suggesting whole lines or blocks of code as you type
- Converting comments into working code
- Helping you understand and work with unfamiliar libraries
- Offering multiple suggestions for you to choose from
- Learning from your coding style and preferences

# Switching AI Models in Copilot

GitHub Copilot offers different AI models to help with various coding tasks. Here's how to switch between them:

## Available Models

- **Default (GPT-3.5 Turbo)**: The standard model for general coding assistance
- **GPT-4**: More powerful model with enhanced code understanding and generation capabilities
- **Copilot Chat**: Conversational interface for complex queries and explanations

## How to Switch Models

1. **In VS Code**:
    - Click on the Copilot icon in the status bar
    - Select "GitHub Copilot: Switch Model"
    - Choose your preferred model from the dropdown menu

2. **Using Command Palette**:
    - Press `Ctrl+Shift+P` (Windows/Linux) or `Cmd+Shift+P` (Mac)
    - Type "GitHub Copilot: Switch Model"
    - Select the desired model

3. **In Copilot Chat**:
    - Type `/model` followed by the model name
    - Example: `/model gpt-4`

## When to Use Different Models

- **Default Model**: Everyday coding tasks, quick suggestions
- **GPT-4**: Complex algorithms, thorough code explanations, challenging problems
- **Specialized Models**: Domain-specific coding (when available)

Remember that more powerful models might be slower to respond or have usage limitations based on your subscription plan.


### ✏️ Exercise: Write a class for managing patient records
   - Include methods for adding, updating, and retrieving records
   - Implement data validation
   - Add methods for basic statistical analysis

````python
class PatientRecords:
    def __init__(self):
        self.records = []

    def add_record(self, patient_id, name, age, gender, diagnosis):
        """Add a new patient record with validation."""
        if not isinstance(patient_id, int) or patient_id <= 0:
            raise ValueError("Patient ID must be a positive integer.")
        if not isinstance(name, str) or not name.strip():
            raise ValueError("Name must be a non-empty string.")
        if not isinstance(age, int) or age <= 0:
            raise ValueError("Age must be a positive integer.")
        if gender not in ['M', 'F']:
            raise ValueError("Gender must be 'M' or 'F'.")
        if not isinstance(diagnosis, str) or not diagnosis.strip():
            raise ValueError("Diagnosis must be a non-empty string.")

        self.records.append({
            "Patient ID": patient_id,
            "Name": name,
            "Age": age,
            "Gender": gender,
            "Diagnosis": diagnosis
        })

    def update_record(self, patient_id, **kwargs):
        """Update an existing patient record."""
        for record in self.records:
            if record["Patient ID"] == patient_id:
                for key, value in kwargs.items():
                    if key in record:
                        record[key] = value
                return
        raise ValueError(f"No record found for Patient ID {patient_id}.")

    def retrieve_record(self, patient_id):
        """Retrieve a patient record by ID."""
        for record in self.records:
            if record["Patient ID"] == patient_id:
                return record
        raise ValueError(f"No record found for Patient ID {patient_id}.")

    def average_age(self):
        """Calculate the average age of patients."""
        if not self.records:
            return 0
        return sum(record["Age"] for record in self.records) / len(self.records)

    def gender_distribution(self):
        """Calculate the distribution of genders."""
        distribution = {"M": 0, "F": 0}
        for record in self.records:
            if record["Gender"] in distribution:
                distribution[record["Gender"]] += 1
        return distribution

    def total_patients(self):
        """Return the total number of patients."""
        return len(self.records)
    
    def sample_data(self):
        """Generate sample patient data."""
        self.add_record(1, "Alice", 28, "F", "Flu")
        self.add_record(2, "Bob", 34, "M", "Diabetes")
        self.add_record(3, "Charlie", 45, "M", "Asthma")
        self.add_record(4, "Diana", 35, "F", "Hypertension")
        self.add_record(5, "Eve", 29, "F", "Allergy")

        # Print out the sample data
        print("Sample Data:")
        for record in self.records:
            print(record)

        # Print total patients
        print("\nTotal Patients:", self.total_patients())

        # Print average age
        print("Average Age:", self.average_age())

        # Print gender distribution
        print("Gender Distribution:", self.gender_distribution())
````

````python
import matplotlib.pyplot as plt
import numpy as np
from matplotlib.widgets import RadioButtons

# Generate sample data
x = np.linspace(0, 10, 100)
y1 = np.sin(x)
y2 = np.cos(x)
y3 = x ** 2

# Create a figure and axis
fig, ax = plt.subplots()
plt.subplots_adjust(left=0.3)  # Adjust space for the interactive elements

# Plot the initial data (default is sine)
line, = ax.plot(x, y1, label="Sine")
ax.set_xlabel("X-axis")
ax.set_ylabel("Y-axis")
ax.legend()

# Add interactive radio buttons to switch between plots
ax_radio = plt.axes([0.05, 0.4, 0.2, 0.4])  # Position for the radio buttons
radio = RadioButtons(ax_radio, ('Sine', 'Cosine', 'Quadratic'))

# Function to update the plot based on the selected option
def update_plot(label):
    if label == 'Sine':
        line.set_ydata(y1)
        ax.set_title("Sine Wave")
    elif label == 'Cosine':
        line.set_ydata(y2)
        ax.set_title("Cosine Wave")
    elif label == 'Quadratic':
        line.set_ydata(y3)
        ax.set_title("Quadratic Function")
    ax.relim()  # Recalculate limits
    ax.autoscale_view()  # Autoscale the view
    fig.canvas.draw_idle()  # Redraw the canvas

# Connect the radio buttons to the update function
radio.on_clicked(update_plot)
# Create an interactable scatter plot
scatter, = ax.plot(x, y1, 'o', label="Scatter")

# Function to update the scatter plot based on the selected option
def update_scatter(label):
    if label == 'Sine':
        scatter.set_ydata(y1)
        ax.set_title("Sine Scatter")
    elif label == 'Cosine':
        scatter.set_ydata(y2)
        ax.set_title("Cosine Scatter")
    elif label == 'Quadratic':
        scatter.set_ydata(y3)
        ax.set_title("Quadratic Scatter")
    ax.relim()  # Recalculate limits
    ax.autoscale_view()  # Autoscale the view
    fig.canvas.draw_idle()  # Redraw the canvas
    # Create a bar graph
    bar = ax.bar(x, y1, label="Bar")

    # Function to update the bar graph based on the selected option
    def update_bar(label):
        if label == 'Sine':
            for rect, height in zip(bar, y1):
                rect.set_height(height)
            ax.set_title("Sine Bar Graph")
        elif label == 'Cosine':
            for rect, height in zip(bar, y2):
                rect.set_height(height)
            ax.set_title("Cosine Bar Graph")
        elif label == 'Quadratic':
            for rect, height in zip(bar, y3):
                rect.set_height(height)
            ax.set_title("Quadratic Bar Graph")
        ax.relim()  # Recalculate limits
        ax.autoscale_view()  # Autoscale the view
        fig.canvas.draw_idle()  # Redraw the canvas

    # Connect the radio buttons to the bar update function
    radio.on_clicked(update_bar)
# Connect the radio buttons to the scatter update function
radio.on_clicked(update_scatter)
# Create a bar graph
bar = ax.bar(x, y1, label="Bar")

# Function to update the bar graph based on the selected option
def update_bar(label):
    if label == 'Sine':
        for rect, height in zip(bar, y1):
            rect.set_height(height)
        ax.set_title("Sine Bar Graph")
    elif label == 'Cosine':
        for rect, height in zip(bar, y2):
            rect.set_height(height)
        ax.set_title("Cosine Bar Graph")
    elif label == 'Quadratic':
        for rect, height in zip(bar, y3):
            rect.set_height(height)
        ax.set_title("Quadratic Bar Graph")
    ax.relim()  # Recalculate limits
    ax.autoscale_view()  # Autoscale the view
    fig.canvas.draw_idle()  # Redraw the canvas

# Connect the radio buttons to the bar update function
radio.on_clicked(update_bar)
plt.show()
````
